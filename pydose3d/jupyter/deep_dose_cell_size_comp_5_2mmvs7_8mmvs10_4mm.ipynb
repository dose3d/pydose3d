```python
import seaborn as sns
import plotly as plot
import pydose3d
import pandas as pd
import matplotlib.pyplot as plt
import pydose3d.data as data
from pydose3d.svc.ntuple_data import NTupleDataSvc
from pydose3d.svc.dose3d import Dose3DSvc
NTupleDataSvc.ImplicitMT = True
pydose3d.set_log_level("INFO")
```



Assumpptions: 
* Different size of scintilator are simulated.

```python

data = [f"/home/g4rt/test/g4rt/output/Size_10_4/10_4_mm_cells/10_4_mm_cells.root"]
```

```python
d3dsvc = Dose3DSvc()
d3dsvc.set_data(data,"Dose3DVoxelisedTTree") #This is only tree name (ROOT TTree) in the patient definition used.
df1 = d3dsvc.get_cell_dose_z_profile_pdframe( MLayer = 0, MColumn = 0, normalized=True)
df2 = d3dsvc.get_cell_dose_z_profile_pdframe( MLayer = 0, MColumn = 1, normalized=True)
df3 = d3dsvc.get_cell_dose_z_profile_pdframe( MLayer = 1, MColumn = 0, normalized=True)
df4 = d3dsvc.get_cell_dose_z_profile_pdframe( MLayer = 1, MColumn = 1, normalized=True)
df1 = df1[["CellPositionZ","Dose"]]
df2 = df2[["CellPositionZ","Dose"]]
df3 = df3[["CellPositionZ","Dose"]]
df4 = df4[["CellPositionZ","Dose"]]
```

```python
def pop_std(x):
    return x.std(ddof=0)


fig, ax = plt.subplots()
plt.rcParams['figure.figsize'] = [18, 8]
plt.rcParams['figure.dpi'] = 100 # 200 e.g. is really fine, but slower
```

```python
df1 = df1.rename(columns={"CellPositionZ":"Z [mm]"})
df2 = df2.rename(columns={"CellPositionZ":"Z [mm]"})
df3 = df3.rename(columns={"CellPositionZ":"Z [mm]"})
df4 = df4.rename(columns={"CellPositionZ":"Z [mm]"})

pdd1 = pd.concat([df1, df2, df3, df4], ignore_index=True) 
pdd1 = pdd1.groupby("Z [mm]").agg({"Dose":['mean',pop_std]}).reset_index()
pdd1.columns = ['Z [mm]','Dose','Std Dev']
pdd1 = pdd1.reindex(columns=sorted(pdd1.columns))
```

```python
sns.scatterplot(data=pdd1,x='Z [mm]',y='Dose')
```
